```python
import yfinance as yf
import pandas as pd
import numpy as np

seel = yf.download(tickers="SEEL", period="7d", interval="5m")

prev = seel["Close"][0]
vol = []
for price in seel["Close"][1:]:
    vol.append(price - prev)
    prev = price
avg = np.mean(vol)
std = np.std(vol)
indices = set()
for index, change in enumerate(vol):
    if change <= avg - 2*std or change >= avg + 2*std:
        indices.add(index + 1)
```

```python
zyne = yf.download(tickers="ZYNE", period="7d", interval="5m")

prev = zyne["Close"][0]
vol = []
for price in zyne["Close"][1:]:
    vol.append(price - prev)
    prev = price
avg = np.mean(vol)
std = np.std(vol)
for index, change in enumerate(vol):
    if change <= avg - 2*std or change >= avg + 2*std:
        neighbor = set()
        for i in range(10):
            neighbor.add((index - 5) + i)
        inter = indices.intersection(neighbor)
        if len(inter) > 0:
            for i in inter:
                print("SEEL")
                print(seel.iloc[i])
                print("ZYNE")
                print(zyne.iloc[index])
```
